The sentiment analysis, performed using NLTK's VADER, provides four metrics for each review:

sentiment_neg: Represents the proportion of negative words in the review.
sentiment_neu: Represents the proportion of neutral words in the review.
sentiment_pos: Represents the proportion of positive words in the review.
sentiment_compound: A normalized, weighted composite score ranging from -1 (most negative) to +1 (most positive), which summarizes the overall sentiment.
Let's break down their distributions and interdependencies:

Distributions of Sentiment Values:
From the descriptive statistics (reviews[['sentiment_neg', 'sentiment_neu', 'sentiment_pos', 'sentiment_compound']].describe()):

Negative Sentiment (sentiment_neg):

The mean negative score is very low (0.047), and the median is even lower (0.026), indicating that most reviews contain very few negative words. A significant portion of reviews have a sentiment_neg of 0, meaning no negative words were detected. The maximum is 0.258, showing a few reviews are quite negative.
Visualized: The histogram for sentiment_neg would show a distribution heavily skewed towards zero, with a long tail extending to higher values for a few truly negative reviews.
Neutral Sentiment (sentiment_neu):

The mean neutral score is around 0.75, with a median of 0.76. This suggests that a large proportion of the words in the reviews are considered neutral. The range is quite narrow (min 0.421, max 0.982), indicating most reviews have a substantial neutral component.
Visualized: The histogram for sentiment_neu would likely be more centrally peaked around 0.7-0.8, reflecting a consistent presence of neutral language.
Positive Sentiment (sentiment_pos):

The mean positive score is 0.20, with a median of 0.19. This is significantly higher than the negative scores, indicating a general positive leaning in the reviews. The maximum is 0.579.
Visualized: The histogram for sentiment_pos would show a distribution skewed towards lower positive values but with a clear presence of higher positive scores, more spread out than the negative scores.
Compound Sentiment (sentiment_compound):

The mean compound score is 0.56, and the median is 0.86. This strong positive median suggests that the overall sentiment of the reviews is overwhelmingly positive. The scores range from -0.9776 (highly negative) to 0.9955 (highly positive), but the 25th percentile (0.44) is already quite positive.
Visualized: The compound score histogram would likely show a bimodal or heavily skewed distribution, with a large peak towards the very positive end (+1) and a smaller, broader distribution covering neutral and negative sentiments.
Interdependencies:
Inverse Relationship between Positive and Negative: While not perfectly inverse (due to the neutral component), reviews with higher sentiment_pos scores tend to have lower sentiment_neg scores, and vice-versa. This is intuitively expected, as a review is typically either positive or negative, or largely neutral.
Neutrality as a Baseline: The sentiment_neu score often accounts for the largest portion of a review's sentiment breakdown. This means that even in positive or negative reviews, a significant percentage of words are not explicitly classified as emotional. The sum of sentiment_neg, sentiment_neu, and sentiment_pos for a given review generally adds up to 1 (or very close to it, excluding minor floating point differences).
Compound Score as a Synthesis: The sentiment_compound score provides an overall summary, taking into account all three (neg, neu, pos) components and their intensity. A high compound score typically indicates high sentiment_pos and low sentiment_neg, whereas a low compound score indicates the opposite.

Topic coherence measures how interpretable the topics generated by an LDA model are. A higher coherence score indicates more meaningful and human-interpretable topics.

```python
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt
import numpy as np
from nltk.corpus import stopwords

# Ensure 'corpus' is defined (from reviews['snippet'].dropna())
# Ensure 'count_vect' is fitted and 'feature_names' are available from previous steps

# Create Gensim dictionary from the CountVectorizer's vocabulary
id2word = Dictionary()
id2word.id2token = {i: word for i, word in enumerate(feature_names)}
id2word.token2id = {word: i for i, word in enumerate(feature_names)}

# Create Gensim corpus (list of lists of (token_id, count)) from x_counts sparse matrix
gensim_corpus = []
for doc_idx in range(x_counts.shape[0]):
    row = x_counts.getrow(doc_idx)
    bow = []
    for col_idx, count in zip(row.indices, row.data):
        bow.append((col_idx, int(count))) # Ensure counts are integers
    gensim_corpus.append(bow)

# Prepare tokenized texts for coherence model (c_v requires original tokenized texts)
# Filter out stopwords and custom stopwords consistent with CountVectorizer's preprocessing
combined_stopwords = set(stopwords.words('english') + mystopwords)
tokenized_texts = []
for text in corpus:
    tokens = count_vect.build_analyzer()(text) # Tokenize and lowercase
    filtered_tokens = [token for token in tokens if token not in combined_stopwords]
    tokenized_texts.append(filtered_tokens)

# Function to compute coherence scores for different numbers of topics
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=1, random_state=100):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit + 1, step): # limit + 1 to include the limit in the range
        print(f"Training LDA model with {num_topics} topics...")
        model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=10, random_state=random_state)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
    return model_list, coherence_values

# Define the range of topics to evaluate
start_topics = 2
limit_topics = 15 # Evaluate up to 15 topics
step_topics = 1

# Compute coherence values for the specified range of topics
model_list, coherence_values = compute_coherence_values(dictionary=id2word,
                                                      corpus=gensim_corpus,
                                                      texts=tokenized_texts,
                                                      start=start_topics,
                                                      limit=limit_topics,
                                                      step=step_topics)

# Plotting the coherence scores
x = range(start_topics, limit_topics + 1, step_topics)
plt.figure(figsize=(10, 6))
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.title("Coherence Score vs. Number of Topics for Gensim LDA")
plt.xticks(x)
plt.grid(True)
plt.show()

# Find the index of the maximum coherence value and suggest the optimal number of topics
optimal_idx = np.argmax(coherence_values)
optimal_num_topics = x[optimal_idx]
print(f"\nSuggested optimal number of topics based on coherence score: {optimal_num_topics}")

# The optimal_lda_model is stored in model_list[optimal_idx] if needed for further analysis
```

```python
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

tickets = [
    {"id": 101, "text": "I love the product! It arrived early and works great."},
    {"id": 102, "text": "The box was crushed and the support team hasn't emailed me back. I am furious!!"},
    {"id": 103, "text": "How do I change my password? I can't find the button."},
    {"id": 104, "text": "This is the worst experience ever. I want a refund IMMEDIATELY 😡"}
]

def triage_tickets(ticket_list):
    print(f"{'ID':<6} | {'SCORE':<8} | {'PRIORITY':<10} | {'PREVIEW'}")
    print("-" * 60)

    for ticket in ticket_list:
        score = analyzer.polarity_scores(ticket['text'])['compound']

        # If score is below -0.5, it's a "Crisis" priority
        if score <= -0.5:
            priority = "🚨 CRISIS"
        elif score < 0:
            priority = "⚠️ HIGH"
        else:
            priority = "✅ NORMAL"

        print(f"{ticket['id']:<6} | {score:<8.2f} | {priority:<10} | {ticket['text'][:40]}...")

triage_tickets(tickets)
```

VADER Sentiment Analysis
I used VASER (Valence Aware Dictionary and Sentiment Reasoner) for short and informal texts like emails and chats in companies.
How it works;
•	Looks for words with certain sentiment strengths like furious, worst or love.
•	It accounts for common writing pattern like intensifiers, punctuation emphasis and negation.
•	It produces a score for every text called compounds.

The code installs and imports the library (Vader sentiment), Creates a sentiment analyzer (analyzer = SentimentIntensityAlalyzer), It calculates the sentiment, then uses numeric scores to assign a priority level;
	score <= -0.5 → 🚨 CRISIS
	score < 0 (but not <= -0.5) → ⚠️ HIGH
	score >= 0 → ✅ NORMAL
The project basically helps in deciding which ticket or sentiment need urgent attention based on how negative the customer language is. \
The results are given in the following ways;
•	Crisis which means that the text is strongly negative, so it needs urgent attention.
•	High means the ticket is negative, but it is not urgent
•	Normal means the ticket is neutral so it can follow the standard procedure.
